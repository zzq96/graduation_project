```python
import tensorflow
import matplotlib.pyplot as plt
import pickle
import  xlrd
import os
import numpy as np
from PIL import Image,ImageFont,ImageDraw
import xlwt
```

```python
#得到汉字和字符的映射
def get_dict(path):
    """
    :param path:存储汉字的excel绝对路径，一列存 
    :return: id2word，word2id
    """
    table = xlrd.open_workbook(path)
    sheet = table.sheet_by_index(0)
    id2word = {}
    word2id = {}
    for i in range(sheet.nrows):
        word = sheet.row_values(i)[0]
        id2word[i] = word
        word2id[word] = i
# print(word, type(word))
    return id2word,word2id

table_path = r'汉字编码字符集一级字库3755个（一列）.xls'
id2word, word2id = get_dict(table_path)
with open(r"Data/id2word.pkl","wb") as f:
    pickle.dump(id2word, f)
with open(r"Data/word2id.pkl","wb") as f:
    pickle.dump(word2id, f)
```

```python
#得到初始数据集
class Font2Img(object):
    """
    
    """
    def __init__(self,
                 width,
                 height,
                 ):
        self.width = width
        self.height = height
    
    def get_image(self,font_path, word, rotate):
        """
        :param font_path: str,字体文件绝对路径 
        :param word: str,需要转化为图像的一个汉字
        :return: narray,size=(height,width),字体图像
        """
        #白色背景
        img = Image.new("RGB", (self.width, self.height), "white")
        draw = ImageDraw.Draw(img)
        font = ImageFont.truetype(font_path, int(self.width*0.7))
        #黑色字体
        draw.text((0, 0), word, (0, 0, 0),font = font)   
        if rotate != 0:
            img = img.rotate(rotate)
        # Image._show(img.rotate(90))
        #转化为narray类型
        img = np.asarray(list(img.getdata()), dtype='uint8').reshape((width, height ,3))
        up, down ,left ,right = self.find_box(img)
        # print(up, down, left, right)
        img = img[up:down+1, left:right+1,:]
        return img
    def find_box(self, img):
        col_sum = np.sum(img[:,:,0], axis = 0)
        row_sum = np.sum(img[:,:,0], axis = 1)
        max_width = 255 * self.width
        max_height = 255 * self.height
        left = 0
        right = 0
        up = 0
        down = 0
        #寻找字体上边界
        for i in range(self.height):
            if row_sum[i] < max_width:
                up = i
                break
        #寻找字体下边界
        for i in range(self.height - 1, -1, -1):
            if row_sum[i] < max_width:
                down = i
                break
         #寻找字体左边界
        for i in range(self.width):
            if col_sum[i] < max_height:
                left = i
                break
         #寻找字体右边界
        for i in range(self.width - 1, -1, -1):
            if col_sum[i] < max_height:
                right = i
                break
        return up, down, left, right
                 

id2word = {}
word2id = {}
with open(r"Data/id2word.pkl","rb") as f:
    id2word = pickle.load(f)
with open(r"Data/word2id.pkl","rb") as f:
    word2id = pickle.load(f)
    
# for item in 
width = 160
height = 160
Rotate = 20
Rotate_step = 5 
# for word, id in word2id.items():
#     print(word, id)
#     font2img = Font2Img(width, height)
#     for font_name in os.listdir(r"./Fonts"):    
#         font_path = os.path.join(r"./Fonts",font_name)
#         for rotate in range(-Rotate, Rotate+1, Rotate_step):
#             img = font2img.get_image(font_path, word, rotate)
            # plt.imshow(img)
            # plt.show() 
```
